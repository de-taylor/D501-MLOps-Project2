# EDA for census.csv

Exploring the csv file using a Pandas dataframe

```python
import os
import pandas as pd
```

```python
if os.path.exists('../data/census.csv') and os.access('../data/census.csv', os.R_OK):
    census_orig = pd.read_csv('../data/census.csv')
else:
    print(os.getcwd())
```

```python
cat_features = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]
label = census_orig['salary']

cat_df = census_orig[cat_features]
cont_df = census_orig.drop(columns=cat_features).drop(columns='salary')
```

```python
for col_name in cat_df.columns:
    print(f"Column: {col_name}")
    # print(cat_df[col_name].value_counts().index)
    print(cat_df[col_name].value_counts().index.values)
```
